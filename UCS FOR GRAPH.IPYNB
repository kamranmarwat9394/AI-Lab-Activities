```python
import heapq

def find_least_cost_path(graph, start, goal):
    # Priority queue stores (cost, current_node, path)
    queue = [(0, start, [start])]
    visited = set()

    while queue:
        (cost, node, path) = heapq.heappop(queue)

        if node in visited:
            continue

        visited.add(node)

        # If we reached the destination, return the result
        if node == goal:
            return cost, path

        # Check neighbors
        for neighbor, weight in graph.get(node, {}).items():
            if neighbor not in visited:
                heapq.heappush(queue, (cost + weight, neighbor, path + [neighbor]))

    return float("inf"), []

# Define your graph based on the connections provided
graph = {
    'S': {'A': 3, 'B': 2, 'C': 7},
    'A': {'D': 3, 'G': 15}, # Assuming the '8' was a typo or part of D/G
    'B': {'G': 20},
    'C': {'G': 6},
    'D': {},
    'G': {}
}

# Execution
cost, path = find_least_cost_path(graph, 'S', 'G')

print(f"Least Cost: {cost}")
print(f"Path taken: {' -> '.join(path)}")
```
